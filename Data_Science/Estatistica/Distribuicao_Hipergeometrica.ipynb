```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
from scipy.special import binom
```

```python
k=np.linspace(0,10,101)
```

```python
K1 = 5
K2 = 8
```

```python
N1 = 40
N2 = 50
```

```python
n1 = 18
n2 = 25
```

```python
bin1 = (binom(K1,k)*binom(N1-K1,n1-k))/binom(N1,n1)
bin2 = (binom(K1,k)*binom(N1-K1,n2-k))/binom(N1,n2)
bin3 = (binom(K2,k)*binom(N1-K2,n1-k))/binom(N1,n1)
bin4 = (binom(K2,k)*binom(N1-K2,n2-k))/binom(N1,n2)
bin5 = (binom(K1,k)*binom(N2-K1,n1-k))/binom(N2,n1)
bin6 = (binom(K1,k)*binom(N2-K1,n2-k))/binom(N2,n2)
```

```python
plt.plot(k,bin1,label='K='+str(K1)+' N='+str(N1)+' n='+str(n1))
plt.plot(k,bin2,label='K='+str(K1)+' N='+str(N1)+' n='+str(n2))
plt.plot(k,bin3,label='K='+str(K2)+' N='+str(N1)+' n='+str(n1))
plt.plot(k,bin4,label='K='+str(K2)+' N='+str(N1)+' n='+str(n2))
plt.plot(k,bin5,label='K='+str(K1)+' N='+str(N2)+' n='+str(n1))
plt.plot(k,bin6,label='K='+str(K1)+' N='+str(N2)+' n='+str(n2))
plt.xlabel('k')
plt.ylabel('P(N,K,n,k)')
plt.legend()
plt.tight_layout()
```

```python
from scipy.stats import hypergeom
```
