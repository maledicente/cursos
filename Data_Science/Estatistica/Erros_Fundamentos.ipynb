```python
import numpy as np
```

```python
valores_reais = np.array([1.1,1.5,2.4,2.9,3.1,3.5,3.7,4.2,4.5,4.9,5.5,5.9,6.2,6.7,7.5,8.5,8.7,8.9])
```

```python
valores_previstos = np.array([1.13,1.46,2.394,3.011,3.114,3.522,3.689,4.154,4.566,5.001,5.545,5.993,6.194,6.712,7.554,8.434,8.721,8.995])
```

# Erro absoluto médio

```python
erros = np.abs(valores_reais - valores_previstos)
```

```python
print(erros)
```

```python
soma_erros = np.sum(erros)
```

```python
print(soma_erros)
```

```python
MAE = soma_erros/len(valores_reais)
```

```python
print("MAE = {:0.3f}".format(MAE))
```

# Erro quadrático médio

```python
erro_quadratico = (np.array(valores_reais) - np.array(valores_previstos))**2.0
```

```python
erro_quadratico
```

```python
soma_erro_quadratico = np.sum(erro_quadratico)
```

```python
soma_erro_quadratico
```

```python
MSE = soma_erro_quadratico/len(valores_reais)
```

```python
print("MSE = {:0.4f}".format(MSE))
```

# Raiz do erro quadrático médio

```python
RMSE = np.sqrt(MSE)
```

```python
print("RMSE = {:0.3f}".format(RMSE))
```

# Erro logaritmico quadrático médio

```python
erro_log_quadratico = (np.log(valores_reais+1)-np.log(valores_previstos+1))**2.0
```

```python
soma_erro_log_quadratico = np.sum(erro_log_quadratico)
```

```python
MSLE = soma_erro_log_quadratico/len(valores_reais)
```

```python
print("MSLE = {:0.5f}".format(MSLE))
```

# Raiz do erro logaritmico quadrático médio

```python
RMSLE = np.sqrt(MSLE)
```

```python
print("RMSLE = {:0.3f}".format(RMSLE))
```

# Erro absoluto percentual médio

```python
soma_erro_percentual = np.sum(np.abs(erros)/valores_previstos)
```

```python
MAPE = soma_erro_percentual*100/len(valores_reais)
```

```python
print("MAPE = {:0.2f}%".format(MAPE))
```
