```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
X = np.linspace(0,5,51)
```

```python
k = [1,2,3,4,5]
```

```python
colors = ['black','red','blue','green','orange']
```

```python
fs = []

for i in range(len(k)):
    f = ((X**(k[i]-1))*np.exp(-X**2.0/2.0))/((2.0**(k[i]/2.0-1))*math.gamma(k[i]/2))
    fs.append(f)
```

```python
for i in range(len(k)):
    plt.plot(X,fs[i],color=colors[i],label='k='+str(k[i]))

plt.xlabel('X')
plt.ylabel('f(X,k)')
plt.legend()
plt.tight_layout()
```
