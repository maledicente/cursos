# Matplotlib - Gráfico de torta

* Criando gráficos de torta com a biblioteca MatPlotLib


* Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

* Criando labels

```python
labels = ["A1","A2","A3","A4","A5"]
```

* Criando lista com valores

```python
valores = [10,45,32,56,23]
```

* Criando gráfico de barras

```python
plt.pie(valores,labels=labels)
plt.show()
```

* Inserindo porcentagem

```python
plt.pie(valores,labels=labels, autopct='%1.1f%%')
plt.show()
```

* Explode

```python
explode = [0,0.1,0,0,0]
```

```python
plt.pie(valores,labels=labels, autopct='%1.1f%%',explode=explode)
plt.show()
```
