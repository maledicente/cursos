# Matplotlib - Legenda

* Manipulando a legenda


* Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

* Criando variável X

```python
X = np.linspace(0,5,11)
```

* Criando funções Y(X) e Z(X)

```python
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

* Colocando equações na legenda

```python
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.plot(X,Z,label='$Z(X)=e^{(x/2)}$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```

* Alterando posição da legenda

```python
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.plot(X,Z,label='$Z(X)=e^{(x/2)}$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend(loc='lower right')
plt.show()
```

* Criando novas variáveis X, Y e Z

```python
X = np.linspace(0,2.0*np.pi,361)
```

```python
Y = np.cos(X/2)
Z = np.sin(X/4)
```

```python
plt.plot(X,Y,label='$\cos(X/2)$',color='red')
plt.plot(X,Z,label='$\sin(X/4)$',color='green',linestyle='--')
plt.legend()
```
