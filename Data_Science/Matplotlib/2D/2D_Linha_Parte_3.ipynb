# Matplotlib - Gráfico de linha - Parte 3

* Parte 3: Tipo da linha


* Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

* Criando variável X

```python
X = np.linspace(0,5,11)
```

* Criando funções Y(X) e Z(X)

```python
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

* Gráfico de Y e Z em função de X

```python
plt.plot(X,Y,label='Y(X)')
plt.plot(X,Z,label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

* Alterando linha da função Z(X)

```python
plt.plot(X,Y,label='Y(X)')
plt.plot(X,Z,label='Z(X)',linestyle='--')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

* Usando nomes para os tipos de linhas

```python
plt.plot(X,Y,label='Y(X)', linestyle='dashed')
plt.plot(X,Z,label='Z(X)',linestyle='dashdot')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
plt.plot(X,Y,label='Y(X)', linestyle='solid')
plt.plot(X,Z,label='Z(X)',linestyle='dotted')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```
