# Matplotlib - Gráfico de linha - Parte 3

* Parte 4: Adicionando preenchimento no gráfico


* Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

* Criando variável X

```python
X = np.linspace(0,2.0*np.pi,360)
```

* Criando funções Y(X) e Z(X)

```python
Y = np.sin(X)
```

```python
Z = np.cos(X)
```

* Gráfico de Y e Z em função de X

```python
plt.plot(X,Y,label='Y(X)')
plt.plot(X,Z,label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```

* Adicionando preenchimento

```python
plt.plot(X,Y,label='Y(X)')
plt.fill_between(X,0,np.sin(X))
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```

```python
plt.plot(X,Y,label='Y(X)')
plt.plot(X,Z,label='Z(X)')
plt.fill_between(X,0,np.sin(X))
plt.fill_between(X,0,np.cos(X))
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```
