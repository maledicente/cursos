# Matplotlib - Gráfico de linha - Parte 2

* Parte 2: Manipulando eixos dos gráficos


* Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

* Criando variável X

```python
X = np.linspace(0,5,11)
```

* Criando funções Y(X) e Z(X)

```python
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

* Gráfico de Y e Z em função de X

```python
plt.plot(X,Y,label='Y(X)')
plt.plot(X,Z,label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

* Alterando eixo X

```python
plt.plot(X,Y,label='Y(X)')
plt.plot(X,Z,label='Z(X)')
plt.xlim([0,2])
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

* Alteranco eixo Y

```python
plt.plot(X,Y,label='Y(X)')
plt.plot(X,Z,label='Z(X)')
plt.xlim([0,2])
plt.ylim([0,5])
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

* Invertendo eixo X

```python
plt.plot(X,Y,label='Y(X)')
plt.plot(X,Z,label='Z(X)')
plt.xlim([2,0])
plt.ylim([0,5])
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

* Rotacionando valores do eixo X

```python
plt.plot(X,Y,label='Y(X)')
plt.plot(X,Z,label='Z(X)')
plt.xlim([2,0])
plt.ylim([0,5])
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.xticks(rotation=45)
plt.show()
```
