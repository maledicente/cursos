# Matplotlib - Gráfico de espalhamento - Parte 3

* Parte 3: Controlando tamanho do ponto


* Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

* Criando variável X

```python
X = np.linspace(0,5,11)
```

* Criando funções Y(X) e Z(X)

```python
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

* Gráfico de Y em função de X

```python
plt.scatter(X,Y)
plt.show()
```

* Alterando tamanho do ponto

```python
plt.scatter(X,Y,s=10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
plt.scatter(X,Y,s=10,label='Y(X)')
plt.scatter(X,Z,s=40,label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```

```python
plt.scatter(X,Y,s=10,label='Y(X)',alpha=0.8)
plt.scatter(X,Z,s=40,label='Z(X)',alpha=0.4)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```
