# Matplotlib - Gráfico de espalhamento - Parte 1

* Parte 1: Criando gráficos de espalhamento (scatter) com a biblioteca MatPlotLib


* Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

* Criando variável X

```python
X = np.linspace(0,5,11)
```

* Criando funções Y(X) e Z(X)

```python
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

* Gráfico de Y em função de X

```python
plt.scatter(X,Y)
plt.show()
```

* Inserindo nomes nos eixos

```python
plt.scatter(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

* Inserindo título no gráfico

```python
plt.scatter(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.show()
```

* Dois gráficos na mesma imagem

```python
plt.scatter(X,Y)
plt.scatter(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

* Alterando cores

```python
plt.scatter(X,Y,color='black')
plt.scatter(X,Z,color='red')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

* Inserindo legenda

```python
plt.scatter(X,Y,color='black',label='Y(X)')
plt.scatter(X,Z,color='red',label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```
