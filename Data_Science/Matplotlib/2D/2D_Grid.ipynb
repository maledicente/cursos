# Matplotlib - Legenda

* Manipulando a grade (malha, grid)


* Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

* Criando variável X

```python
X = np.linspace(0,5,11)
```

* Criando funções Y(X) e Z(X)

```python
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

* Criando grid

```python
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid(b=True , linestyle = '-' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```

* Alterando cor do fundo

```python
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid(linestyle = '--' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```
