# Matplotlib - Gráfico de superfície 3D

* Criando gráficos de scatter 3D com a biblioteca MatPlotLib


* Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

* Criando variáveis X e Y

```python
X = np.linspace(-5,5,11)
Y = np.linspace(-5,5,11)
```

* Convertendo variáveis X e Y para malha

```python
X, Y = np.meshgrid(X, Y)
```

* Criando variável Z

```python
Z = X**2 + Y**2
```

* Gráfico 3D

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,Z)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')
plt.tight_layout()
plt.show()
```

* Usando mapa de cores

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X,Y,Z, cmap=cm.coolwarm, antialiased=False)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')
plt.tight_layout()
plt.show()
```

* Barra de cores

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X,Y,Z, cmap=cm.coolwarm, antialiased=False)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.tight_layout()
plt.show()
```
