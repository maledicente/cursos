# Matplotlib - Gráfico de linha 3D

* Criando gráficos de linha 3D com a biblioteca MatPlotLib


* Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

* Criando variáveis X e Y

```python
X = np.linspace(-5,5,11)
Y = np.linspace(-5,5,11)
```

* Criando variável Z

```python
Z1 = 2*X**2 - 5*Y**2
Z2 = -2*X**2 + 8*Y**3
```

* Gráfico 3D

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import axes3d
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(X,Y,Z1)
plt.show()
```

* Inserindo nomes nos eixos

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(X,Y,Z1)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')
plt.tight_layout()
plt.show()
```

* Inserindo dois gráficos

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(X,Y,Z1,label='Z1')
ax.plot(X,Y,Z2,label='Z2')
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_zlabel('Eixo Z')
plt.legend()
plt.tight_layout()
plt.show()
```
